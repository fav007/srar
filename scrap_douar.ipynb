```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import pandas as pd

# Initialize the WebDriver
driver = webdriver.Firefox()  # or webdriver.Firefox()

# Navigate to the login page
driver.get('https://172.31.254.24/gndar/www/index.php?module=jauth&action=login:form')

# Fill in the login form
username = driver.find_element(By.NAME, 'login')
password = driver.find_element(By.NAME, 'password')

username.send_keys('user')
password.send_keys('password')

# Submit the form
login_button = driver.find_element(By.XPATH, '//*[@id="valider-acces"]')
login_button.click()

# Wait for the login to complete (you may need to adjust this)
driver.implicitly_wait(2)  # Wait for 10 seconds
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'html.parser')
links = soup.find_all('a')
print(links)
driver.find_element(By.LINK_TEXT, "Traiter un BSC" ).click()
time.sleep(10)
driver.find_element(By.CLASS_NAME,'linkU').click()
driver.find_element(By.ID, "ui-accordion-accordion-header-4").click()
driver.find_elements(By.CLASS_NAME,'linkU')[0].click()
driver.find_elements(By.CLASS_NAME,'linkU')[1].click()
driver.find_elements(By.CLASS_NAME,'linkU')[2].click()
driver.find_elements(By.CLASS_NAME,'linkU')[3].click()
```

```python
table = driver.find_element(By.ID,"ui-accordion-accordion-panel-7")
soup = BeautifulSoup(table.get_attribute('outerHTML'), "html.parser")
table_headers = []
for th in soup.find_all('th'):
    table_headers.append(th.text)

table_data = []
for row in soup.find_all('tr'):
    columns = row.find_all('td')
    output_row = []
    for column in columns:
        output_row.append(column.text)
    table_data.append(output_row)
df = pd.DataFrame(table_data, columns=table_headers).dropna().apply(lambda x:x.str.strip())
df
```
